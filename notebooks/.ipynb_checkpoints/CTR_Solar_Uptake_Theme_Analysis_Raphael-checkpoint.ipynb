```python
import pandas as pd
import numpy as np
import os 
import seaborn as sns
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
def vars_to_themes(vars_list, themes_set, grp_data):
    temp = []
    themes = {}
    for x in themes_set:
        temp = [y for y in grp_data.get_group(x).Var_Name.values]
        themes.setdefault(x, temp[:])
    return themes

def font_color(val):
    color = 'white' if (val > 0.1) else 'black'
    return 'color: %s' % color
```

# Costs Vs. Targets

```python
vz = sns.heatmap(costs_mtx, annot=False, fmt='.1g', cmap='Spectral',linecolor='k',linewidth=.1)
vz.axes.set_title("Costs",fontsize=15)
vz.set_xlabel("Target Variables",fontsize=10)
vz.set_ylabel("Cost Variables",fontsize=10)
vz.figure.dpi=100
vz.yaxis.set_tick_params(labelsize=10)
vz.xaxis.set_tick_params(labelsize=10)
vz.figure.set_figwidth(10)
```

```python
index= costs_mtx.index
cols = costs_mtx.columns
costs_mtx = pd.DataFrame(costs_mtx.values, index=index, columns=cols)
costs_mtx.style.applymap(font_color).background_gradient().format('{:.2%}')
```
